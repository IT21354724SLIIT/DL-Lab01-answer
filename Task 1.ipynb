### 3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5]. 


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Create the figure
fig = plt.figure(figsize=(12, 8))

# Create a 3D axis
ax = Axes3D(fig)

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0.1)

# Add a color bar
fig.colorbar(surf, shrink=0.5, aspect=5)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z = X^2 + Y^2')

plt.show()
```

#### 4. Calculate the pearson standard coefficient and spearman rank correlation for the [HP, Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the results using heatmaps. Make sure the values are displaced in the heatmap as well.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select the relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Calculate Pearson correlation
pearson_corr = df[features].corr(method='pearson')

# Calculate Spearman rank correlation
spearman_corr = df[features].corr(method='spearman')

# Set up the matplotlib figure

# Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Heatmap')
```

```python
# Create Spearman correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Rank Correlation Heatmap')

plt.show()
```
